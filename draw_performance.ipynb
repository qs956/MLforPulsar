```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data

```python
single=np.load('../result/performance/performance_single.npy')
two=np.load('../result/performance/performance_two.npy')
```

# Setting

```python
titlesize = 18
fontsize = 16
```

# Draw

```python
clf_name=['DT','Adaboost','Xgboost','Gboost']
```

```python
#single
x=range(1,23)
for i in range(4):#4种算法
    for j in range(4):#4种采样
        if (j==1 or j==2):
            continue;
        y=[]
        for k in range(22):#每种采样22个特征
            y.append(single[i*88+j*22+k][0])
        plt.figure(figsize=(8,5))
        plt.plot(x,y,'*--',label='Recall')
        y=[]
        for k in range(22):#每种采样22个特征
            y.append(single[i*88+j*22+k][1])
        plt.plot(x,y,'--',label='Precision')
        y=[]
        for k in range(22):#每种采样22个特征
            y.append(single[i*88+j*22+k][2])
        plt.plot(x,y,label='F1 Score')
        plt.plot(x,0.95*np.ones(np.shape(x)),'r--')
        plt.legend(fontsize=fontsize,loc=1)
        plt.xlim([1,22])
        plt.ylim([0,1])
        plt.xticks(range(1,23),size=fontsize)
        plt.yticks(size=fontsize)
        if(j==0):
            plt.title('Single feature selection for Non-SMOTE '+clf_name[i],size=fontsize)
        if(j==3):
            plt.title('Single feature selection for SMOTE '+clf_name[i],size=fontsize)
        plt.savefig('../figure/1219/'+clf_name[i]+'_'+str(j)+'.png')
```
